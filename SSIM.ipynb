```python
# pillowをインポート
from PIL import Image, ImageFilter
import numpy as np
import random
import matplotlib.pyplot as plt
import math
import heapq
import cv2
from skimage.metrics import structural_similarity as ssim
from skimage import io, color

main_size = 32
b_size = 16

start = 1
end = 24
for kodimnum in range(start, end+1):
    defoname = "IMG"+str(kodimnum).zfill(4)
    
    Lstart = 1
    Lend = 4
    interval = 1
    #print("kodimnum =", kodimnum+1)
    #for L in range(1):
    for L in range(Lstart, Lend+1, interval):
        filename1 = "C:/Users/cesa0/VRAE_LSB/new_kodak_gray_VRAELSB/motomura_color_decrypted/"+defoname+"_motomura_color_b16_L"+str(L)+"_decrypted.tiff"
        filename1 = "C:\\Users\\cesa0\\VRAE_LSB\\SB\\"+str(main_size)+"_"+str(b_size)+"_color_decrypted3\\"+defoname+"_"+str(main_size)+"_"+str(b_size)+"_L"+str(L)+"_color_ver3_decrypted.tiff"
        #filename1 = defoname+"_motomura_L"+str(L)+"_b16_decrypted.tiff"
        #filename1 = defoname+"_VRAELSB_L"+str(L)+"_b16_decrypted.tiff"
        #filename1 = defoname+"_VRAELSB2_L"+str(L)+"_b16_decrypted.tiff"
        #filename = defoname+".tiff"
        #filename1 = "test6.tiff"

        original = io.imread(defoname+'.tiff')
        decrypted1 = io.imread(filename1)

        # カラー画像 → YCbCr変換
        original_ycbcr = color.rgb2ycbcr(original)
        decrypted_ycbcr = color.rgb2ycbcr(decrypted1)

        # 輝度成分(Yチャンネル)を取り出し
        Y_original = original_ycbcr[:, :, 0]
        Y_decrypted = decrypted_ycbcr[:, :, 0]
        
        #print(str(ssim(original, decrypted1)))

        # SSIMを輝度成分で比較
        score = ssim(Y_original, Y_decrypted, data_range=Y_decrypted.max() - Y_decrypted.min())
        
        print(score)
```
